```python
# We adopt the above code and implement a filtering system that updates an empty dictionary with the game's details
gamedict ={}
gameDatalist = totalGameInfo[2].splitlines()
for index, datum in enumerate(gameDatalist): #iterate each index or line
    line = datum.split() #array of strings for the current index or line
    key = '' #dictionary key
    Lstring = ' ' #empty string used for updating
    
    position = [] 
    for index2, string in enumerate(line): #iterate through the string array of current index or line
        if ':' in string: #find the position of ':'
            position.append(index2)
 
    for index2, string in enumerate(line): #iterate through current index with position of ':' known first
        if index == 0: #if the first line
            gamedict.update({'Title:':" ".join(line)}) #update the title with the first line with strings joined together
        
        elif index == 1:
            gamedict.update({'appid:':int(" ".join(line))})
            
        elif (string == 'Site') or (index == 2) or (string == 'Store') or (string == 'Free'):
            if index == 2:
                gamedict.update({'Description:':" ".join(line)})
            elif line[0] == 'Free':
                gamedict.update({'Price:':'$0'})
        
        elif (string == 'Developer:') or (string == 'Publisher:'):
            newlabel1 = string
            value1 = " ".join(line[position[0]+1:position[1]])
            gamedict.update({newlabel1:value1})
        
        elif (string == 'Owners:') or (string == 'Price:') or (string == 'Followers:') or (string == 'Metascore:'):
            newlabel1 = string
            value1 = line[index2+1]
            gamedict.update({newlabel1:value1})
            
        elif string == 'Owners':
            newlabel1 = " ".join(line[index2:index2+4])
            value1 = line[index2+4]
            gamedict.update({newlabel1:value1})
    
        elif (string == 'Old'):
            newlabel1 = " ".join(line[index2:index2+2])
            value1 = line[index2+2]
            gamedict.update({newlabel1:value1})
        
        elif (string == 'Peak') or (string == 'Playtime') or (string == 'Release') or (string == 'YouTube'):
            newlabel1 = " ".join(line[:position[0]+1])
            value1 = " ".join(line[position[0]+1:])
            gamedict.update({newlabel1:value1})
            
        elif (string == 'Genre:') or (string == 'Languages:') or (string == 'Tags:') or (string == 'Category:'):
            newlabel1 = string
            value1 = " ".join(line[position[0]+1:])
            gamedict.update({newlabel1:value1})
```

```python
# Below we make the above filtering system into a function that can be used on each game string
def string_to_dictionary(gamestring):
    gamedict ={}
    gameDatalist = gamestring.splitlines()
    for index, datum in enumerate(gameDatalist):
        line = datum.split()
        key = ''
        Lstring = ' '
        position = []
        for index2, string in enumerate(line):
            if ':' in string:
                position.append(index2)
 
        for index2, string in enumerate(line):
            try:
                if index == 0:
                    gamedict.update({'Title:':" ".join(line)})

                elif index == 1:
                    gamedict.update({'appid:':int(" ".join(line))})

                elif (string == 'Site') or (index == 2) or (string == 'Store') or (string == 'Free'):
                    if index == 2:
                        gamedict.update({'Description:':" ".join(line)})
                    elif line[0] == 'Free':
                        gamedict.update({'Price:':['$0']})

                elif (string == 'Developer:') or (string == 'Publisher:'):
                    newlabel1 = string
                    value1 = " ".join(line[position[0]+1:position[1]])
                    gamedict.update({newlabel1:value1})

                elif (string == 'Owners:') or (string == 'Price:') or (string == 'Followers:') or (string == 'Metascore:'):
                    newlabel1 = string
                    if string == 'Price:':
                        value1 = line[index2+1:]
                    else:
                        value1 = line[index2+1]
                    gamedict.update({newlabel1:value1})

                elif string == 'Owners':
                    newlabel1 = " ".join(line[index2:index2+4])
                    value1 = line[index2+4]
                    gamedict.update({newlabel1:value1})

                elif (string == 'userscore:'):
                    newlabel1 = 'Old userscore:'
                    value1 = line[index2+1]
                    gamedict.update({newlabel1:value1})

                elif (string == 'Peak') or (string == 'Playtime') or (string == 'Release') or (string == 'YouTube'):
                    newlabel1 = " ".join(line[:position[0]+1])
                    value1 = " ".join(line[position[0]+1:])
                    gamedict.update({newlabel1:value1})

                elif (string == 'Genre:') or (string == 'Languages:') or (string == 'Tags:') or (string == 'Category:'):
                    newlabel1 = string
                    value1 = " ".join(line[position[0]+1:])
                    gamedict.update({newlabel1:value1})
            except:
                print('Error at this index: ', index2, ' And this string: ', string)

    return gamedict 
```

```python
# We use the above filtering system in a defined function
def final_price(priceArray):
    ''' This takes in the price of the game originally formated as a string array and returns a float price value. 
        If there is a discount by the price, then the discount is used to calculate the original cost of the price.
        
        Input array must not have NaN.
    '''
    if len(priceArray) == 2: #price has a discount next to it
        newpriceformat = []
        for string in priceArray:
            newpriceformat.append(re.sub('[$,%]', '', string)) 
        originalprice = round(float(newpriceformat[0])/((100-int(newpriceformat[1]))/100), 2)
    elif len(priceArray) == 1: #no discount next to price
        originalprice = round(float(re.sub('[$,]', '', priceArray[0])), 2)
    else:
        print('Price not converted or handled correctly')
    if originalprice >= 0: #no error with the discount
    #Note above we converted free games to have a value of '$0'
        return originalprice
    elif originalprice <0: #The discount was incorrect and made the price wrong
        return round(float(re.sub('[$,]', '', priceArray[0])), 2) #just returns the discounted price
        
```

```python
# We filter out all the NaN from the price data column
df[df['Price:'].isna() == False]['Price:'].head(50)
```

```python
# We implment the filter above so we can use the defined function for cleaning the price data
df['Price:'] =  df[df['Price:'].isna() == False]['Price:'].map(lambda x: final_price(x))
df['Price:']
```
